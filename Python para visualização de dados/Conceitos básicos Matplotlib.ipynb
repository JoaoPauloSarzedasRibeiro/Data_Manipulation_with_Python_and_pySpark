```python
import matplotlib.pyplot as plt
```

```python
#pra mostrar os gráficos na linha (funciona no Jupyter - pra outros teriam que usar o plt.show())
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x*x
```

```python
y
```

```python
#Forma funcional
plt.plot(x,y, 'r--')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Título')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g--')
```

```python
fig = plt.figure()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.45,0.8])
axes2 = fig.add_axes([0.65,0.1,0.45,0.8])
#10% do canto esquerdo, 10% do canto inferior, 80% de largura e Altura
axes1.plot(x,y, 'r--')
axes1.set_xlabel('Eixo x')
axes1.set_title('Título 1')
axes2.plot(y,x, 'g')
axes2.set_xlabel('Eixo x')
axes2.set_title('Título 2')
```

```python
#cria a figura e ja cria os eixos ou multiplos eixos
fig, ax = plt.subplots()
#acessando um método da instancia de classe eixo e não mais o plt.plot
ax.plot(x, x**3,'b--')
ax.plot(x, x**4,'g--')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Título');
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
for axe in ax:
    axe.plot(x,y,'b')
    axe.set_title('Título')
fig
```

```python
#assim começo a trabalhar usando metodos e instancias de uma classe,
#o que me dá muito mais controle sobre o layout
ax[0].plot(x,y,'b')
ax[0].set_title('Título')

ax[1].plot(x,y,'b')
ax[1].set_title('Título')

fig
```

```python
fig, ax = plt.subplots(nrows=5, ncols=5)
plt.tight_layout() #evita sobreposição
```

```python
fig = plt.figure(figsize=(0,4),dpi=100)
```

```python
fig,axes = plt.subplots(figsize=(10,5))
axes.plot(x,y,'r--')
axes.set_title('Título')
```

```python
fig,axes = plt.subplots(figsize=(6,8))
axes.plot(x,y,'r--', label='x^2')
axes.plot(y,x,'g--', label='Dados 2')
axes.set_title('Título')
axes.legend(loc=0)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x,x**2,'g-.')
ax.plot(x,x**3,color='#880088', linewidth=1, alpha=0.5, linestyle='--', marker='o', markersize=5) 
#alpha é transparencia

ax.set_xlim([0,2]) #limites de valores do eixo x
ax.set_ylim([0,10]) #limites do eixo y
```
